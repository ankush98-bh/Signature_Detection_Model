```python
import os
import numpy as np
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
from google.colab.patches import cv2_imshow
from PIL import Image
```

```python
class sign_detection():
  def __init__(self,with_sign_files,without_sign_files,test_path):
    self.with_sign_files=with_sign_files
    self.without_sign_files=without_sign_files
    self.test_path=test_path

  def preprocessing_img(self):
    x=os.listdir(self.with_sign_files)
    y=os.listdir(self.without_sign_files)
    print(len(x))
    print(len(y))

    self.with_sign_labels=[1]*len(x)
    self.without_sign_labels=[0]*len(y)
    self.labels=self.with_sign_labels+self.without_sign_labels
    print(len(self.with_sign_labels))
    print(len(self.without_sign_labels))

    with_sign_path = self.with_sign_files
    without_sign_path=self.without_sign_files
    self.data = []
    for filename in os.listdir(with_sign_path):
      image=Image.open(with_sign_path+filename)
      image=image.resize((128,128))
      image=image.convert('RGB')
      image=np.array(image)
      image=image/255.0
      self.data.append(image)

    for filename in os.listdir(without_sign_path):
      image=Image.open(without_sign_path+filename)
      image=image.resize((128,128))
      image=image.convert('RGB')
      image=np.array(image)
      image=image/255.0
      self.data.append(image)
    print(len(self.data))
    print(len(self.labels))

  def Nueral_network(self):
    self.model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(512, activation='relu'),
    Dense(units=1, activation='sigmoid')  # Binary classification (signature or not)
    ])

    self.model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
    print("Model Build Succsesfully")

  def training_model(self):
    x=np.array(self.data)
    y=np.array(self.labels)
    print(len(x))
    print(len(y))

    # Split data into training and validation sets
    self.x_train, self.x_test, self.y_train, self.y_test = train_test_split(x, y, test_size=0.2, random_state=42)

    # Train the model
    history=self.model.fit(self.x_train, self.y_train,validation_split=0.2,epochs=10)

  def evalution(self):
    loss, accuracy = self.model.evaluate(self.x_test, self.y_test)
    print(f'Validation loss: {loss}, Validation accuracy: {accuracy}')

  def testing(self):
    test_image = load_img(self.test_path, target_size=(128, 128))
    test_image_array = img_to_array(test_image) / 255.0
    # Predict using the trained model
    prediction = self.model.predict(np.array([test_image_array]))
    print(prediction)
    # Display the image and its prediction
    #plt.imshow(test_image)
    if prediction > 0.5:
        plt.title('Congratulation..! You got your Signature')
        image_data_crop=cv2.imread(test_image)
        hight,width=image_data_crop.shape[:2]

        #start_row,sart_col= int(hight*.0.20),int(width*0.20)   #extracted the pixle cordinate (starting from top left)
        start_row,start_col=int(hight*0.20),int(width*0.20)
        end_row,end_col=int(hight*.80),int(width*.80) #extracted the pixle cordinate (ending to bottom right)

        cropped_image=image_data_crop[start_row:end_row,start_col:end_col]
        cv2.imshow(cropped_image)
        cv2.waitkeys(0)
        cv2.destroyAllWindows()

    else:
        plt.title('Sorry..! Signature not present in image')
    #plt.show()
```

```python
import os
import numpy as np
import cv2
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Dropout, UpSampling2D, concatenate

class SignatureSegmentation:
    def __init__(self, image_size=(256, 256)):
        self.image_size = image_size
        self.model = self.build_model()

    def build_model(self):
        inputs = Input(shape=(self.image_size[0], self.image_size[1], 3))

        # Encoding path
        conv1 = Conv2D(32, (3, 3), activation='relu', padding='same')(inputs)
        conv1 = Conv2D(32, (3, 3), activation='relu', padding='same')(conv1)
        pool1 = MaxPooling2D(pool_size=(2, 2))(conv1)

        # Decoding path
        conv5 = Conv2D(64, (3, 3), activation='relu', padding='same')(pool1)
        conv5 = Conv2D(64, (3, 3), activation='relu', padding='same')(conv5)
        up6 = UpSampling2D((2, 2))(conv5)
        concat6 = concatenate([conv1, up6], axis=3)
        conv6 = Conv2D(32, (3, 3), activation='relu', padding='same')(concat6)
        conv6 = Conv2D(32, (3, 3), activation='relu', padding='same')(conv6)

        # Output layer
        outputs = Conv2D(1, (1, 1), activation='sigmoid')(conv6)

        model = Model(inputs=[inputs], outputs=[outputs])
        model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
        return model

    def preprocess_image(self, image_path):
        image = cv2.imread(image_path)
        image = cv2.resize(image, self.image_size)
        image = image / 255.0
        return image

    def train(self, data_dir, batch_size=4, epochs=10):
        train_data_gen = tf.keras.preprocessing.image.ImageDataGenerator(
            rescale=1./255, validation_split=0.2)

        train_generator = train_data_gen.flow_from_directory(
            data_dir,
            target_size=self.image_size,
            batch_size=batch_size,
            class_mode='input',
            subset='training'
        )

        validation_generator = train_data_gen.flow_from_directory(
            data_dir,
            target_size=self.image_size,
            batch_size=batch_size,
            class_mode='input',
            subset='validation'
        )

        self.model.fit(train_generator, validation_data=validation_generator, epochs=10)

    def segment_signature(self, input_image_path, output_mask_path):
        input_image = self.preprocess_image(input_image_path)
        prediction = self.model.predict(np.expand_dims(input_image, axis=0))[0]

        # Threshold the prediction to create a binary mask
        threshold = 0.5
        binary_mask = (prediction > threshold).astype(np.uint8) * 255

        # Save the binary mask as an output image
        cv2.imwrite(output_mask_path, binary_mask)

# Example usage
signature_segmentation = SignatureSegmentation()
signature_segmentation.train('/content/drive/MyDrive/Ankush_AINXT/Augmented_data_with_sign')

input_image_path = '/content/drive/MyDrive/Ankush_AINXT/With_sign_data/data (10).jpg'
output_mask_path = '/content/drive/MyDrive/Ankush_AINXT/With_sign_data/data (10).jpg'
signature_segmentation.segment_signature(input_image_path, output_mask_path)
```
